```python
# importing the libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# importing the yield df

zoom = pd.read_csv('yield.csv')
zoom.tail()
```

```python
# the values is renamed to the desired name

zoom = zoom.rename(index=str, columns={"Value": "hg/ha_yield"})
zoom.head()
```

```python
# drop columns with less info

zoom = zoom.drop(['Year Code','Element Code','Element','Year Code','Area Code','Domain Code','Domain','Unit','Item Code'], axis=1)
zoom.head()
```

```python
# visualizing the rain df

rain = pd.read_csv('rainfall.csv')
rain.tail()
```

```python
# renamed the area and also converted an object to float

rain = rain.rename(index=str, columns={" Area": 'Area'})
rain['average_rain_fall_mm_per_year'] = pd.to_numeric(rain['average_rain_fall_mm_per_year'],errors = 'coerce')
```

```python
# merged yield df and rain df

yield_df = pd.merge(zoom, rain)
yield_df.tail()
```

```python
# visualized the pesticides df

df_pes = pd.read_csv('pesticides.csv')
df_pes.head()
```

```python
# drop columns with less info and renamed the value to pesticides tonnes

df_pes = df_pes.rename(index=str, columns={"Value": "pesticides_tonnes"})
df_pes = df_pes.drop(['Element','Domain','Unit','Item'], axis=1)
df_pes.head()
```

```python
# combined the yield_df with the pesticides df to give yield_df

yield_df = pd.merge(yield_df, df_pes)
yield_df.head()
```

```python
# visualize the temp df

avg_temp = pd.read_csv('temp.csv')
avg_temp.head()
```

```python
# converted the country columns to area

avg_temp = avg_temp.rename(index=str, columns={"year": "Year", "country":'Area'})
avg_temp.head()
```

```python
# merged the whole datasets together

yield_df = pd.merge(yield_df,avg_temp, on=['Area','Year'])
yield_df.head()
```

```python
# eliminated the null values

yield_df['average_rain_fall_mm_per_year'].fillna(int(yield_df['average_rain_fall_mm_per_year'].mean()), inplace=True)
```

```python
yield_df.isnull().sum()
```

```python
# encoding categorical variables and defining the input and output data

yield_df_onehot = pd.get_dummies(yield_df, columns=['Area',"Item"])
x=yield_df_onehot.loc[:, yield_df_onehot.columns != 'hg/ha_yield']
y=yield_df['hg/ha_yield']
x.head()
```

```python
x = x.drop(['Year'], axis=1)
x.head()
```

```python
# bringing all the inputs to same magnitued

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
x
```

```python
# splitting the data

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=101, test_size=0.2)
```

```python
# creating the model

from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(x_train, y_train)
```

```python
# making predictions

y_pred = model.predict(x_test)
y_pred
```

```python
# evaluate the model

from sklearn.metrics import r2_score
r2_score(y_pred, y_test)
```
